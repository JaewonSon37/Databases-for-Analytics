### Visualization in Python

```python
import numpy
import matplotlib.pyplot as plt

from numpy.random import randn
```

```python
# Create a figure object
fig = plt.figure()
```

```python
# Add first subplot in a 2x2 grid at position 1
sp1 = fig.add_subplot(2, 2, 1)
randn(30) # Generate 30 random numbers
sp1.plot(randn(30))
fig.set_size_inches(15, 10)
fig
```

```python
# Add second subplot in a 2x2 grid at position 2
sp2 = fig.add_subplot(2, 2, 2)
sp2.scatter(randn(20), randn(20), color = 'r')
sp2.scatter(randn(20), randn(20), color = 'g')
sp2.scatter(randn(20), randn(20), color = 'b')
fig
```

```python
# Add third subplot in a 2x2 grid at position 3
sp3 = fig.add_subplot(2, 2, 3)
sp3.hist(randn(40), bins = 10, color = 'g')
fig
```

```python
# Add fourth subplot in a 2x2 grid at position 4
sp4 = fig.add_subplot(2, 2, 4)
t = numpy.arange(0, 3.0, 0.01) # Generate an array of values for the x-axis
s = numpy.cos(2 * numpy.pi * t) # Compute the cosine values for each x value
sp4.plot(t, s)
sp4.annotate('Center!', xy = (1.5, 0), xytext = (1.2, 0.5)) # Annotate the subplot with a label
fig
```

```python
# Overlay additional dashed line plots
plt.plot(randn(20), color = 'purple', linestyle = 'dashed')
plt.plot(randn(20), color = 'purple', linestyle = 'dashed', marker = 'o')
```

```python
# Create a figure object
figx = plt.figure()
subfig = figx.add_subplot(1, 1, 1)
```

```python
# Generate random data
data = randn(20)
```

```python
subfig.plot(data, 'k--', label = 'Initial Plot')
figx
```

```python
# Plot data with a post-step drawing style
subfig.plot(data, 'k-', drawstyle = 'steps-post', label = 'Post-Step Drawing Style')
subfig.legend(loc = 'lower left') # Add a legend
figx
```

```python
# Plot data with an offset and pre-step drawing style
subfig.plot(data + 2, 'k-.', drawstyle = 'steps-pre', label = 'Pre-Step Drawing Style', color = 'g', linewidth = 5)
figx
```
